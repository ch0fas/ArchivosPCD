## 5. Estructura del prompt (versi√≥n clara y pr√°ctica) üí¨

El *prompt* es la l√≠nea que te muestra el shell cuando est√° listo para recibir comandos. Suele incluir:

- **Host / m√°quina** (p. ej. `MAQUINA`)

### ¬øQu√© es SSH?
SSH (Secure Shell) es un protocolo seguro para conectarse a otros ordenadores o servicios usando cifrado. Para GitHub, SSH permite autenticarte con un par de claves (privada/p√∫blica) y evitar ingresar usuario/contrase√±a en cada `git push`.

### Generar clave (recomiendo ed25519):
```bash
ssh-keygen -t ed25519 -C "tu_email@ejemplo.com"
```
### A√±adir la clave al ssh-agent:
```bash
eval "$(ssh-agent -s)"
ssh-add ~/.ssh/id_ed25519
```
### Copiar la clave p√∫blica al portapapeles (Git Bash):
```bash
clip < ~/.ssh/id_ed25519.pub
```
Pega la clave en GitHub: Settings ‚Üí SSH and GPG keys ‚Üí New SSH key.

### Probar la conexi√≥n:
```bash
ssh -T git@github.com
```
Salida esperada: `Hi tu_usuario! You've successfully authenticated, but GitHub does not provide shell access.`

### Archivo `~/.ssh/config` (opcional):
```text
Host github.com
  HostName github.com
  User git
  IdentityFile ~/.ssh/id_ed25519
  IdentitiesOnly yes
```
