# üìë Gesti√≥n de dependencias con `requirements.txt`

El archivo `requirements.txt` es la forma **cl√°sica y m√°s simple** de manejar dependencias en proyectos de Python.

Sirve para **listar las librer√≠as necesarias** para ejecutar un proyecto y, opcionalmente, fijar versiones espec√≠ficas.

El archivo `requirements.txt` es un documento que lista todas las bibliotecas y paquetes de Python que tu proyecto necesita para funcionar correctamente. Es especialmente √∫til cuando quieres compartir tu proyecto con otras personas o cuando lo quieres desplegar en otro entorno (por ejemplo, en un servidor).
---

## üîπ Ejemplo de `requirements.txt`

```txt
numpy==1.26.0
pandas>=2.0.0,<3.0.0
scikit-learn
matplotlib
```

- `numpy==1.26.0` ‚Üí versi√≥n exacta.

- `scikit-learn` ‚Üí instalar√° la √∫ltima versi√≥n disponible.

# üöÄ Manejadores modernos de dependencias

Ya vimos que `requirements.txt` es la forma **cl√°sica** de gestionar dependencias en Python.
Sin embargo, en proyectos m√°s grandes o colaborativos, este enfoque tiene **limitaciones**:

- No administra entornos virtuales por s√≠ mismo.
- No asegura una **resoluci√≥n consistente** de versiones en distintos sistemas.
- Es dif√≠cil manejar dependencias de desarrollo vs dependencias de producci√≥n.
- No integra metadatos del proyecto (nombre, versi√≥n, autor, etc.).

üëâ Por eso surgieron herramientas **modernas de gesti√≥n de dependencias** que extienden y mejoran el flujo de trabajo.

---

## üîπ Ventajas de los manejadores modernos

- Manejan **dependencias y entornos virtuales** en un solo flujo.
- Generan **archivos declarativos** (`pyproject.toml`, `Pipfile`) y **archivos de bloqueo** (`.lock`) para garantizar reproducibilidad.
- Permiten definir la **versi√≥n de Python** compatible con el proyecto.
- Mejoran la **colaboraci√≥n** en equipos y despliegues en producci√≥n.
- Algunos permiten **publicar paquetes** en PyPI f√°cilmente (ej. Poetry, Hatch).

---

## üîπ Principales herramientas

A continuaci√≥n, un resumen de los m√°s utilizados en la industria:

### üì¶ pip + pip-tools
- **Descripci√≥n:** la base de todo, con `pip-tools` como extensi√≥n para resolver dependencias y generar un `requirements.txt` m√°s confiable.
- **Ventajas:** simple, estable, ampliamente soportado.
- **Desventajas:** m√°s manual, no tiene integraci√≥n de proyecto.
- **Sitio oficial:** [https://pip.pypa.io](https://pip.pypa.io)

---

### üì¶ Pipenv
- **Descripci√≥n:** combina manejo de entornos virtuales y dependencias usando `Pipfile` y `Pipfile.lock`.
- **Ventajas:** flujo integrado, reemplazo natural de `requirements.txt`.
- **Desventajas:** menos popular hoy, resoluci√≥n de dependencias lenta en proyectos grandes.
- **Sitio oficial:** [https://pipenv.pypa.io](https://pipenv.pypa.io)

---

### üì¶ Poetry
- **Descripci√≥n:** muy popular en proyectos modernos, usa `pyproject.toml` y `poetry.lock`. Facilita publicaci√≥n de paquetes.
- **Ventajas:** gran balance entre simplicidad y poder.
- **Desventajas:** la resoluci√≥n de dependencias puede ser lenta.
- **Sitio oficial:** [https://python-poetry.org](https://python-poetry.org)

---

### üì¶ PDM
- **Descripci√≥n:** gestor ligero que implementa PEPs modernos (usa `pyproject.toml`).
- **Ventajas:** r√°pido, m√°s minimalista que Poetry.
- **Desventajas:** menos comunidad y adopci√≥n.
- **Sitio oficial:** [https://pdm.fming.dev](https://pdm.fming.dev)

---

### üì¶ Hatch
- **Descripci√≥n:** flexible, dise√±ado para proyectos con m√∫ltiples entornos y automatizaci√≥n.
- **Ventajas:** muy √∫til para librer√≠as y proyectos de Python avanzado.
- **Desventajas:** menos usado en ciencia de datos.
- **Sitio oficial:** [https://hatch.pypa.io](https://hatch.pypa.io)

---

### üì¶ Rye
- **Descripci√≥n:** administrador moderno de Python + dependencias, enfocado en productividad.
- **Ventajas:** administra versiones de Python, r√°pido, sigue los √∫ltimos PEPs.
- **Desventajas:** m√°s reciente, comunidad peque√±a.
- **Sitio oficial:** [https://rye-up.com](https://rye-up.com)

---

### ‚ö° uv
- **Descripci√≥n:** gestor ultrarr√°pido, pensado como reemplazo de `pip` y `Poetry`. Administra Python, dependencias y entornos.
- **Ventajas:** velocidad ‚ö°, soporta reproducibilidad con `uv.lock`.
- **Desventajas:** a√∫n muy nuevo, documentaci√≥n limitada.
- **Sitio oficial:** [https://docs.astral.sh/uv](https://docs.astral.sh/uv)

# üìÇ Archivos de configuraci√≥n y bloqueo

Hasta ahora vimos c√≥mo manejar dependencias con `requirements.txt`.
Sin embargo, los manejadores modernos de dependencias utilizan un **par de archivos clave** que mejoran la reproducibilidad y organizaci√≥n:

---

## 1. `pyproject.toml`
üìå Es un archivo de configuraci√≥n est√°ndar en Python (definido por [PEP 518](https://peps.python.org/pep-0518/))
- Contiene la informaci√≥n del proyecto: nombre, versi√≥n, autor, dependencias, build system, etc.
- Sirve como **fuente √∫nica de la verdad** para configurar tu entorno.
- Diferentes herramientas lo leen: `poetry`, `uv`, `hatch`, `pdm`, etc.

üëâ Ejemplo b√°sico de `pyproject.toml` con `poetry`:
```toml
[tool.poetry]
name = "mi-proyecto"
version = "0.1.0"
description = "Ejemplo de uso de Poetry"
authors = ["Cristian Zapata <cristianz@iteso.mx>"]

[tool.poetry.dependencies]
python = "^3.11"
numpy = "^1.26"
pandas = "^2.0"
```

---

## 2. `poetry.lock` / `uv.lock` / `pdm.lock`
üìå Estos son **archivos de bloqueo de dependencias**.
- Se generan autom√°ticamente cuando instalas dependencias.
- Guardan las **versiones exactas** que se resolvieron en la instalaci√≥n.
- Garantizan que tu proyecto se pueda reproducir en cualquier m√°quina de forma id√©ntica.

üëâ Ejemplo:
- En un equipo se guarda que `numpy==1.26.4` y `pandas==2.1.0`.
- En otro equipo, al usar `poetry install` o `uv sync`, instalar√° **exactamente esas versiones**, evitando inconsistencias.

---

‚úÖ Diferencia con `requirements.txt`:
- `requirements.txt` lo defines t√∫ manualmente.
- `.lock` lo genera la herramienta autom√°ticamente y siempre refleja **el estado real del entorno**.

---

üìå En conclusi√≥n:
- `pyproject.toml` = **definici√≥n de lo que quieres**.
- `.lock` = **fotograf√≠a exacta de lo que tienes instalado**.

# ‚ö° Uso b√°sico de `uv`: entornos, dependencias y reproducibilidad

`uv` es un gestor ultrarr√°pido que simplifica el ciclo: **crear entorno ‚Üí declarar dependencias ‚Üí ejecutar ‚Üí reproducir**.
Aqu√≠ ver√°s los comandos esenciales y el flujo recomendado para clases y equipos.

---

## 1) Crear el entorno virtual

> Desde la carpeta ra√≠z del proyecto (que ya tiene tu c√≥digo y estructura).

```bash
# Crea un entorno virtual en .venv/ (recomendado)
uv venv
```

- Crea `.venv/` con la versi√≥n de Python por defecto del sistema (o la pinnada si existe `.python-version`).
- No necesitas activar manualmente el entorno para usar `uv` (ver `uv run` m√°s abajo).

> üí° Si prefieres activarlo manualmente:
> - Linux/Mac/Git Bash: `source .venv/bin/activate`
> - Windows (PowerShell): `.\\.venv\\Scripts\\Activate.ps1`
> - Desactivar: `deactivate`

---

## 2) Agregar dependencias

```bash
# Agregar dependencias de ejecuci√≥n
uv add numpy pandas

# Agregar dependencias de desarrollo (por ejemplo, linters/test)
uv add --dev pytest black
```

Esto:
1) Declara dependencias en `pyproject.toml`.
2) Genera/actualiza `uv.lock` con **versiones exactas** (reproducibilidad asegurada).

---

## 3) Instalar / sincronizar dependencias existentes

```bash
# Reconstruye el entorno a partir de pyproject.toml + uv.lock
uv sync
```

- Crea `.venv/` si no existe.
- Instala **exactamente** las versiones fijadas en `uv.lock`.

---

## 4) Ejecutar comandos dentro del entorno (sin activar)

```bash
# Ejecutar Python o cualquier comando dentro del entorno
uv run python -V
uv run python src/main.py
uv run pytest
```

`uv run` habilita el entorno s√≥lo para ese comando.

---

## 5) Comandos √∫tiles del d√≠a a d√≠a

```bash
# Listar dependencias instaladas
uv pip list

# Actualizar una dependencia a la √∫ltima compatible y refrescar lockfile
uv add --upgrade pandas

# Quitar una dependencia
uv remove numpy

# Buscar un paquete (atajo a pypi)
uv search "requests"

# Ver y regenerar lockfile (normalmente se gestiona solo)
# uv lock    # (si necesitas forzar una regeneraci√≥n)
```

> Nota: normalmente no editas `uv.lock` a mano. `uv` lo mantiene por ti al agregar/actualizar deps.

# üêç Gesti√≥n de versiones de Python con `uv`

Adem√°s de manejar dependencias y entornos, **`uv` puede instalar y seleccionar versiones de Python** por proyecto sin depender de herramientas externas (como `pyenv`). Esto facilita que cada proyecto use **exactamente** la versi√≥n que necesita.

---

## 1) Ver qu√© tienes e instalar nuevas versiones

### Listar Pythons detectados por `uv`
```bash
uv python list
```

### Buscar versiones disponibles
```bash
uv python find 3.11
uv python find 3.12
```

### Instalar una versi√≥n espec√≠fica
```bash
uv python install 3.11.9
uv python install 3.12.5
```

> `uv` descarga binarios oficiales para tu plataforma y los deja listos para usar.

---

## 2) Fijar la versi√≥n del proyecto (pin)

La manera m√°s limpia de asegurar que un proyecto use una versi√≥n concreta es **pinnarla** (crear `.python-version`):

```bash
uv python pin 3.11.9
```

Esto crea un archivo **`.python-version`** en la ra√≠z del proyecto con el valor `3.11.9`.
Desde ese momento, comandos como `uv venv`, `uv sync` o `uv run` **respetar√°n** esa versi√≥n.

- Ver/confirmar la versi√≥n activa para el proyecto:
```bash
uv python which
```

- Quitar el pin (volver a usar la predeterminada del sistema o la m√°s cercana compatible):
```bash
uv python unpin
```

---

## 3) Crear o sincronizar el entorno con una versi√≥n concreta

Si **no** usas `.python-version`, puedes indicar la versi√≥n puntualmente:

```bash
# Crear el entorno con una versi√≥n espec√≠fica
uv venv --python 3.11.9

# Sincronizar/instalar respetando una versi√≥n
uv sync --python 3.12
```

> Si esa versi√≥n no est√° instalada, `uv` intentar√° descargarla e instalarla autom√°ticamente.

---

## 4) Declarar compatibilidad en `pyproject.toml`

En los proyectos modernos, tambi√©n es buena pr√°ctica **declarar** la versi√≥n de Python en el archivo de proyecto para comunicar compatibilidad:

```toml
[project]
name = "mi-proyecto"
version = "0.1.0"
requires-python = ">=3.11,<3.12"
```

- `requires-python` no instala Python; solo declara el **rango soportado**.
- El **pin** (`.python-version`) o `--python` son los que **seleccionan** la versi√≥n que usar√° el entorno.

---

## 5) Uso diario con `uv run` (sin activar manualmente)

No necesitas activar el entorno para usar la versi√≥n correcta; `uv` lo hace por ti:

```bash
uv run python -V
uv run python src/main.py
```

Si existe `.python-version`, `uv` usar√° esa versi√≥n. Si no existe, usar√° la indicada con `--python` o la disponible m√°s adecuada.
