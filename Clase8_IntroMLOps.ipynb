En esta clase abordaremos Machine Learning Operations (MLOps), un concepto clave para llevar modelos de machine learning a producci√≥n de forma eficaz. Veremos qu√© es MLOps, por qu√© es importante, sus componentes principales, los retos que ayuda a resolver, mejores pr√°cticas para iniciarse, su evoluci√≥n hist√≥rica y los diferentes niveles de madurez (del 0 al 4) en su implementaci√≥n. El objetivo es sentar una base te√≥rica s√≥lida, complementada con ejemplos pr√°cticos y referencias a fuentes reconocidas, para que comprendas c√≥mo MLOps integra la ciencia de datos con las pr√°cticas de ingenier√≠a de software.

## 1. ¬øQu√© es Machine Learning Operations (MLOps)?

https://www.databricks.com/glossary/mlops

MLOps (Machine Learning Operations) es una disciplina colaborativa que une el desarrollo y la operaci√≥n de sistemas de machine learning, aplicando principios de DevOps al ciclo de vida de los modelos ML. En otras palabras, MLOps es la pr√°ctica de tratar los modelos de ML como artefactos de software que deben ser dise√±ados, construidos, desplegados, monitorizados y mantenidos con el mismo rigor que cualquier otro componente de software. Esto incluye la automatizaci√≥n y monitoreo de todos los pasos: preparaci√≥n de datos, entrenamiento, validaci√≥n, despliegue en producci√≥n, tracking de versiones y m√°s.

Algunos puntos clave para entender MLOps:

- **Unificaci√≥n de procesos:** MLOps busca unificar el ciclo de vida de desarrollo de modelos ML con el ciclo de despliegue de aplicaciones de software. Los modelos, datos y pipelines de ML se integran en un proceso de CI/CD conjunto con el c√≥digo de la aplicaci√≥n.
- **Colaboraci√≥n interdisciplinaria:** Involucra a cient√≠ficos de datos, ingenieros de ML, especialistas en DevOps y TI trabajando juntos. Cada rol aporta herramientas y perspectivas diferentes para que los modelos no solo alcancen buen rendimiento en el laboratorio, sino que sean reproducibles, escalables y fiables en producci√≥n
- **Automatizaci√≥n y calidad:** MLOps enfatiza la automatizaci√≥n de pruebas, integraciones y despliegues para evitar errores manuales. Tambi√©n incorpora pr√°cticas de control de versiones, pruebas de datos/modelos y monitoreo continuo, reduciendo la ‚Äúdeuda t√©cnica‚Äù asociada a sistemas de ML mal gestionados

## üåê MLOps y su relaci√≥n con DevOps
En desarrollo de software tradicional, **DevOps** introdujo dos pr√°cticas esenciales:

- **CI (Continuous Integration)**: integraci√≥n continua.
  Cada vez que un desarrollador hace un cambio de c√≥digo, este se integra y valida autom√°ticamente con pruebas unitarias e integrales.
  üëâ Evita que los cambios rompan la aplicaci√≥n.

- **CD (Continuous Delivery/Deployment)**: entrega y despliegue continuo.
  Permite que cada versi√≥n validada de software se despliegue de manera confiable y frecuente en producci√≥n.
  üëâ Reduce el riesgo y acelera la entrega de valor.

---

## üöÄ ¬øC√≥mo se traslada esto a ML?
En MLOps, el ciclo no es solo **c√≥digo**, tambi√©n incluye **datos y modelos**:

- **CI para ML:**
  - Validaci√≥n autom√°tica de cambios en c√≥digo de entrenamiento.
  - Tests de integridad de datos (esquemas, calidad).
  - Validaci√≥n de experimentos (¬øel nuevo modelo supera al anterior?).

- **CD para ML:**
  - Despliegue automatizado de modelos en entornos de *staging* y producci√≥n.
  - Versionado de modelos + rollback r√°pido si falla el nuevo.
  - Estrategias seguras: *blue/green*, *canary*.

‚ö° En algunos contextos se habla tambi√©n de **CT (Continuous Training)**:
pipelines que reentrenan autom√°ticamente el modelo cuando hay **nuevos datos** o se detecta *drift*.


En resumen, MLOps es a la vez una cultura de ingenier√≠a y un conjunto de pr√°cticas cuyo objetivo es llevar modelos de machine learning de la idea a la producci√≥n de manera r√°pida, consistente y confiable.

# 2. ¬øPor qu√© es importante MLOps? (Motivaci√≥n) üöÄ

> **Idea:** entrenar un modelo ‚â† ponerlo en producci√≥n. El 80% del esfuerzo real est√° en **operarlo y mantenerlo**.

## Problemas comunes sin MLOps üòµ‚Äçüí´
- **Brecha notebook ‚Üí producci√≥n:** c√≥digo no estructurado, dependencias difusas y ‚Äúfunciona-en-mi-m√°quina‚Äù.
- **Deuda t√©cnica oculta:** *pipelines* fr√°giles, acoplamientos invisibles, ‚Äúcambiar algo lo cambia todo‚Äù.
- **No reproducibilidad:** resultados distintos al reentrenar por datos/semillas/entornos diferentes.
- **Sin monitoreo:** el modelo se degrada (*drift*, *decay*) y nadie lo detecta a tiempo.
- **Handoff doloroso:** cient√≠ficos de datos ‚Äúpasan‚Äù el modelo; ingenier√≠a reescribe/ajusta bajo presi√≥n.

## Qu√© habilita MLOps ‚úÖ
- **Ciclos r√°pidos a producci√≥n:** CI/CD/CT para pasar de meses ‚Üí d√≠as/horas.
- **Calidad continua:** pruebas de datos/modelo/pipeline antes de promover un *candidate*.
- **Trazabilidad total:** versionado de **c√≥digo + datos + modelo + config** (auditor√≠a y *rollback*).
- **Monitoreo vivo:** m√©tricas de servicio (latencia, errores) y de ML (drift, performance real).
- **Colaboraci√≥n real:** contratos claros entre DS/ML Eng/DevOps/Negocio y entornos reproducibles (contenedores).

## Ejemplo ilustrativo üß©
- **Antes (sin MLOps):**
  Un `model.pkl` sale del notebook por correo. En producci√≥n falla por un *preprocessing* distinto y dependencias rotas. No hay versi√≥n ni *logs* confiables.
- **Despu√©s (con MLOps):**
  Pipeline orquestado reentrena con datos nuevos, valida m√©tricas y esquemas, registra artefactos en *Model Registry*. Si supera *gates*, se despliega en *staging* (canary/blue-green) y se monitorea en tiempo real.

## Beneficios para el negocio üìà
- **Velocidad** (time-to-value): iteraciones m√°s cortas y seguras.
- **Confiabilidad**: menos regresiones, m√°s *uptime*.
- **Escalabilidad**: de 1 a N modelos con procesos repetibles.
- **Cumplimiento**: auditor√≠a y gobernanza integradas desde el dise√±o.

> **Resumen:** MLOps convierte modelos prometedores en **valor sostenible** al integrar automatizaci√≥n, pruebas, trazabilidad y monitoreo de extremo a extremo.


# 3. Componentes principales de MLOps üõ†Ô∏è

Un sistema de *Machine Learning* en producci√≥n es mucho m√°s que un modelo entrenado.
MLOps organiza los **componentes clave** del ciclo de vida de ML para que funcionen como una sola unidad, reproducible y mantenible.

---

## 3.1 Datos üìä
> - **Ingesta:** automatizar la captura de datos (ETL, streaming, APIs).
> - **Validaci√≥n:** asegurar calidad (esquema, valores esperados, outliers).
> - **Versionado:** guardar *snapshots* de datasets para reproducir entrenamientos.
> - **Feature Store:** repositorio central para definir y reutilizar *features*.

üí° *Ejemplo:*
Un pipeline diario descarga transacciones de una base, valida columnas y guarda la versi√≥n `ventas_2025_09_30`.

```python
import pandas as pd
from datetime import date

df = pd.read_csv("ventas.csv")
assert "monto" in df.columns
df.to_parquet(f"ventas_{date.today()}.parquet")
```

---

## 3.2 Modelos ü§ñ
> - **Entrenamiento:** scripts modulares y reproducibles (semillas, config files).
> - **Evaluaci√≥n:** m√©tricas claras (accuracy, AUC, F1, m√©tricas de negocio).
> - **Registro:** guardar modelos y metadatos en un **Model Registry**.
> - **Governance:** revisi√≥n y documentaci√≥n antes de producci√≥n.

üí° *Ejemplo con MLflow:*
 ```python
import mlflow, mlflow.sklearn

mlflow.start_run()
mlflow.log_param("max_depth", 5)
mlflow.log_metric("accuracy", 0.87)
mlflow.sklearn.log_model(model, "modelo")
mlflow.end_run()
```

---

## 3.3 C√≥digo üíª
> - **APIs y servicios:** exponer el modelo como microservicio o funci√≥n.
> - **Pipelines de deployment:** CI/CD adaptados para ML.
> - **Infraestructura reproducible:** contenedores Docker, Kubernetes, serverless.
> - **Pruebas:** unitarias, integraci√≥n y contratos de API.

*Ejemplo:*
Empaquetar modelo + API en un contenedor slim de Python y desplegarlo en Kubernetes.

 ---

## 3.4 Procesos üîÑ
> - **Integraci√≥n continua (CI):** validaci√≥n autom√°tica de c√≥digo y datos.
> - **Entrega continua (CD):** despliegues consistentes y repetibles.
> - **Entrenamiento continuo (CT):** reentrenos autom√°ticos cuando cambian datos.
> - **Monitoreo:** m√©tricas de servicio y m√©tricas del modelo.
> - **Alertas y retraining:** triggers para reentrenar ante *drift*.

---

## üìå Resumen visual

| Componente | Ejemplo de pr√°ctica MLOps |
|------------|----------------------------|
| **Datos** | Ingesta + validaci√≥n + versionado (Feature Store) |
| **Modelos** | Entrenamiento reproducible + registro + governance |
| **C√≥digo** | API del modelo + CI/CD con contenedores |
| **Procesos** | CI + CD + CT + monitoreo + alertas |

---
**Idea :** los modelos no viven solos. Son parte de un **ecosistema** de datos, c√≥digo y procesos que MLOps conecta y automatiza.



# 4. Retos que MLOps ayuda a abordar ‚ö°

Las pr√°cticas de MLOps surgieron para atacar los **dolores frecuentes** en proyectos de ML.
Estos son los principales retos y c√≥mo se mitigan:

---

## 4.1 Reproducibilidad y gesti√≥n de versiones üîÑ
- Dif√≠cil recrear experimentos (datos, c√≥digo, hiperpar√°metros).
- Sin versionado, no se puede auditar un modelo en producci√≥n.
- **MLOps:** versiona *todo* (c√≥digo, datos, configuraciones, modelos).

*Ejemplo con MLflow:*
```python
import mlflow

with mlflow.start_run():
    mlflow.log_param("modelo", "RandomForest")
    mlflow.log_param("n_estimators", 200)
    mlflow.log_metric("accuracy", 0.89)
    mlflow.sklearn.log_model(model, "modelo")
```

## 4.2 Integraci√≥n entre equipos üë•
- *Handoff* doloroso: DS ‚Üí Ing de software ‚Üí Ops.
- Reescritura de c√≥digo, errores de entorno, ‚Äúworks on my notebook‚Äù.
- **MLOps:** interfaces claras, modelos empaquetados (Docker, ONNX), pipelines CI/CD.

---

## 4.3 Pipelines fr√°giles y deuda t√©cnica üï∏Ô∏è
- Dependencias ocultas, c√≥digo ‚Äúpegamento‚Äù, *feedback loops* inadvertidos.
- Cualquier cambio rompe el sistema (*entanglement*).
- **MLOps:** modularizaci√≥n, pruebas unitarias/integraci√≥n, orquestadores (Airflow, Prefect, Kubeflow).

---

## 4.4 Escalabilidad y rendimiento üöÄ
- El modelo funciona en local, pero no escala en producci√≥n.
- **MLOps:** contenedores, Kubernetes, paralelizaci√≥n, GPU/TPU, *batching* de inferencia.

---

## 4.5 Monitoreo y deriva de datos üìâ
- Modelos en producci√≥n pierden precisi√≥n con el tiempo (*drift*).
- Sin monitoreo, la degradaci√≥n pasa desapercibida.
- **MLOps:** m√©tricas online (latencia, error), comparaci√≥n de distribuciones, alertas autom√°ticas.

üí° *Ejemplo de chequeo de drift simple:*
```python
import numpy as np

base_mean = np.mean(x_train["monto"])
current_mean = np.mean(x_live["monto"])

if abs(current_mean - base_mean) > 0.15 * base_mean:
    print("‚ö†Ô∏è Posible data drift detectado")
```

## üìå Resumen
- **Integraci√≥n fluida** ‚Üí CI/CD + contenedores.
- **Pipelines robustos** ‚Üí orquestaci√≥n + testing.
- **Escalabilidad** ‚Üí infraestructura nativa en la nube.
- **Deriva** ‚Üí monitoreo y reentrenamiento autom√°tico.

üëâ **Idea:** MLOps convierte el caos experimental en un **sistema confiable, escalable y trazable**.


# 5. Mejores pr√°cticas iniciales en MLOps üå±

Dar los primeros pasos en MLOps no significa implementar todo de golpe.
Se puede comenzar con pr√°cticas simples que aportan gran valor:

---

## 5.1 Control de versiones para *todo* üóÇÔ∏è
- Versionar c√≥digo, datos, configuraciones y modelos.
- Usar Git/GitHub para c√≥digo + DVC/Git-LFS para datos grandes.
- Registro de modelos con MLflow, SageMaker o Azure ML.

---

## 5.2 Automatiza pruebas y validaciones üß™
- **Datos:** validaci√≥n de esquemas, detecci√≥n de *outliers*.
- **Modelos:** asegurar que supera m√©tricas m√≠nimas antes del despliegue.
- **Integraci√≥n:** tests de API y *contracts* de features.

üí° *Ejemplo con `asserts` simples:*
```python
assert df.isnull().sum().max() == 0, "‚ö†Ô∏è Datos con nulos inesperados"
assert modelo.accuracy > 0.80, "‚ö†Ô∏è Modelo no cumple criterio de calidad"
```

## 5.3 CI/CD adaptado a ML ‚öôÔ∏è
- Pipelines autom√°ticos que integran entrenamiento, validaci√≥n y despliegue.
- Integraci√≥n con herramientas de CI/CD (GitHub Actions, GitLab, Jenkins).
- Entrenamiento continuo (CT) disparado por nuevos datos.

---

## 5.4 Paridad entre desarrollo y producci√≥n üîÑ
- Usar contenedores para reproducir entornos (Docker, Conda, uv).
- Evitar *training-serving skew*: mismo pipeline de features en entrenamiento e inferencia.
- Infraestructura como c√≥digo para entornos reproducibles.

---

## 5.5 Seguimiento de experimentos üìä
- Registrar par√°metros, m√©tricas y artefactos de cada entrenamiento.
- Usar herramientas de experiment tracking (MLflow, W&B, Neptune).
- Comparar candidatos y elegir *champion* para producci√≥n.

---

## 5.6 Monitoreo y alertas tempranas üì°
- Definir m√©tricas clave (latencia, drift, accuracy real).
- Configurar alertas autom√°ticas para degradaci√≥n del modelo.
- Conectar monitoreo con procesos de reentrenamiento.

---

## 5.7 Cultura y documentaci√≥n üìù
- Documentar supuestos y decisiones de dise√±o.
- Fomentar colaboraci√≥n multidisciplinaria.
- Incorporar retroalimentaci√≥n del negocio para mejorar el ciclo.

---

## üìå Resumen
- **Automatiza pipelines** (CI/CD/CT).
- **Reproduce entornos** (contenedores, IaC).
- **Monitorea continuamente**.
- **Documenta y colabora**.

üëâ *Estos pasos consolidan la base para madurar hacia niveles avanzados de MLOps.*

# 6. Evoluci√≥n hist√≥rica del enfoque MLOps üìú

El t√©rmino **MLOps** es relativamente nuevo, pero naci√≥ de la necesidad de llevar modelos de ML a producci√≥n con el mismo rigor que el software tradicional.
Su evoluci√≥n ha pasado por varias etapas:

---

## 6.1 Antes de 2015: ML artesanal üßë‚Äçüî¨
- Modelos entrenados en notebooks y compartidos como archivos (`.pkl`, `.sav`).
- Procesos manuales, sin versionado ni monitoreo.
- ‚ÄúFunciona en mi compu‚Äù era la norma.

---

## 6.2 2015‚Äì2017: Conciencia y primeras herramientas üõ†Ô∏è
- Google publica *‚ÄúHidden Technical Debt in ML Systems‚Äù* (2015), visibilizando los problemas de deuda t√©cnica.
- Surgen frameworks de orquestaci√≥n y experimentaci√≥n:
  - **Airflow** (ETL y pipelines).
  - **TFX (TensorFlow Extended)**.
  - **Kubeflow** (Google, para Kubernetes).
  - **MLflow** (Databricks, para tracking y registro).
- Aparece DVC para versionado de datos.

---

## 6.3 2018: Formalizaci√≥n y nubes ‚òÅÔ∏è
- Grandes *clouds* lanzan servicios integrados:
  - AWS SageMaker.
  - Google AI Platform (hoy Vertex AI).
  - Azure ML.
- Empresas como Uber (Michelangelo) y Facebook (FBLearner Flow) muestran plataformas internas de ML a escala.
- ‚ÄúMLOps‚Äù empieza a usarse formalmente en conferencias y blogs.

---

## 6.4 2019‚Äì2020: Adopci√≥n masiva üöÄ
- Se populariza el t√©rmino y aparecen **modelos de madurez** (Google, Microsoft, Azure).
- CI/CD se adapta al ML: validaci√≥n de datos, pruebas de modelo, *Continuous Training (CT)*.
- Startups de MLOps ofrecen *feature stores*, monitoreo y gobernanza.
- MLOps se vuelve requisito en proyectos serios de IA.

---

## 6.5 2021‚Äì2023: Expansi√≥n y nuevos retos üåç
- **Edge MLOps**: desplegar modelos en m√≥viles/IoT.
- **IA Responsable**: fairness, explicabilidad, sesgo integrados en el pipeline.
- Observabilidad avanzada: conexi√≥n con APM/Logging.
- Nace **LLMOps**: MLOps aplicado a *Large Language Models*.
- Automatizaci√≥n de extremo a extremo: modelos que se reentrenan solos con triggers.

---

## 6.6 Futuro üîÆ
- **Pipelines aut√≥nomos**: autoajuste de hiperpar√°metros, AutoML continuo.
- **ModelOps**: gestionar cientos/miles de modelos simult√°neamente.
- **Est√°ndar de la industria**: como DevOps lo fue para software, MLOps ser√° inevitable para ML.

---

## üìå Resumen
- De **ML artesanal** ‚Üí a **MLOps moderno** en menos de 10 a√±os.
- Impulsado por el dolor real de llevar ML a producci√≥n.
- Habilitado por frameworks open-source y *clouds*.
- Hoy es un campo en expansi√≥n con variantes (Edge MLOps, LLMOps, ModelOps).

üëâ **Idea:** MLOps no es moda; es una **respuesta natural a los problemas reales** de poner ML en producci√≥n.

````python
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
from sklearn.feature_extraction import DictVectorizer
````

````python
df = pd.read_parquet('./data/green_tripdata_2024-01.parquet')

df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

df = df[(df.duration >= 1) & (df.duration <= 60)]
````

````python
categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']

df[categorical] = df[categorical].astype(str)
train_dicts = df[categorical + numerical].to_dict(orient='records')

train_dicts
````

````python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)

target = 'duration'
y_train = df[target].values

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_train)
````

````python
root_mean_squared_error(y_train, y_pred)
````

````python
sns.distplot(y_pred, label='prediction')
sns.distplot(y_train, label='actual')

plt.legend()
````

````python
def read_dataframe(filename):

    df = pd.read_parquet(filename)

    df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

    df = df[(df.duration >= 1) & (df.duration <= 60)]

    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)

    return df
````

````python
df_train = read_dataframe('./data/green_tripdata_2024-01.parquet')
df_val = read_dataframe('./data/green_tripdata_2024-02.parquet')
len(df_train), len(df_val)
````

````python
df_train['PU_DO'] = df_train['PULocationID'] + '_' + df_train['DOLocationID']
df_val['PU_DO'] = df_val['PULocationID'] + '_' + df_val['DOLocationID']

categorical = ['PU_DO']  #'PULocationID', 'DOLocationID']
numerical = ['trip_distance']
````

````python
dv = DictVectorizer()

train_dicts = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
````

````python
target = 'duration'

y_train = df_train[target].values
y_val = df_val[target].values

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

root_mean_squared_error(y_val, y_pred)
````
