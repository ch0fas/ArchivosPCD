```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando datos como CSV

dflinks = pd.read_csv("data/links.csv")
dfmovies = pd.read_csv("data/movies.csv")
dfratings = pd.read_csv("data/ratings.csv")
dftags = pd.read_csv("data/tags.csv")

df = pd.merge(dflinks, dfmovies, how="left", on="movieId")
df = pd.merge(df, dfratings, how="left", on="movieId")
df = pd.merge(df, dftags, how="left", on="movieId")
df
```

# EDA

```python
# What was the average rating for Toy Story?

df_toystory = df[df.title == "Toy Story (1995)"]
df_toystory.mean(axis=0, numeric_only=True) # The average rating was 3.9209 
```
